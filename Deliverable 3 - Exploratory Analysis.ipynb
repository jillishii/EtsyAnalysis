```python
import matplotlib.pyplot as plt
% matplotlib inline
import numpy as np
import seaborn as sns
```

```python
occurances = category_df['top_category'].value_counts()
```

```python
objects = []
for i in range(15):
    objects.append(occurances.index[i])

y_pos = np.arange(len(objects))
plt.bar(y_pos, occurances, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Number of Subcategories')
plt.xlabel('Top Category')
plt.title('Number of Subcategories in each Top Category')
plt.show()
```
